## 36 고차 미분 이외의 용도

### 36.1 double backprop의 용도

```
import numpy as np
from dezero import Variable

x = Variable(np.array(2.0))
y = x**2
y.backward(create_graph=True)
gx = x.grad
x.cleargrad()

z = gx ** 3 + y
z.backward()
print(x.grad)
```
