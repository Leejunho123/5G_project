### 6.1 Variable 클래스 추가 구현

#### Variable 설명
+ data는 그 전처럼 data
+ grad는 역전파를 이용한 각각의 기울기(미분값)

```
class Variable:
  def __init__(self, data) :
    self.data = data
    self.grad = None
```
