```
# Function 변경
class Function:
  def __call__(self, input):
    x = input.data
    y = self.forward(x)
    output = Variable(y)
    return output
  
  def forward(self, x):
    raise NotImplementedError()
```

#### NotImplementedError()
+ 하위 클래스에서 오버라이드를 했을 때 반드시 수정해줘야하는 부분에 주로 사용하며 만약 수정을 안했다면 추가적으로 입력하는 에러 메세지를 출력한다.

```
class Square(Function):
  def forward(self, x):
    return x**2
```

```
x = Variable(np.array(10))
```

```
f = Square()
y = f(x)
print(type(y))
print(y.data)
```
