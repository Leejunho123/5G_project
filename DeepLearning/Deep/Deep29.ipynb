## 29 뉴턴 방법으로 푸는 최적화 (수동 계산)
+ 6회만에 도달가능

### 29.1 뉴턴 방법을 활용한 최적화 이론
+ 2차까지 미분한 테일러 급수를 이용하면 y의 최솟값은 x = a - f'(a)/f''(a) 에 있다는 것

### 29.2 뉴턴 방법을 활용한 최적화 구현

```
import numpy as np 
from dezero import Variable

def f(x):
  y = x ** 4 - 2 * x ** 2
  return y

def gx2(x):
  return 12 * x ** 2 - 4

x = Variable(np.array(2.0))
iters = 10

for i in range(iters):
  print(i, x)

  y = f(x)
  x.cleargrad()
  y.backward()

  x.data -= x.grad / gx2(x.data)
```
